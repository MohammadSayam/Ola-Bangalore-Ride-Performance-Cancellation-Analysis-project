```python
text_cols = [
    'Booking Status','Vehicle Type',
    'Pickup Location','Drop Location'
]

for col in text_cols:
    df[col] = df[col].str.strip().str.title()
```

```python
df['Ride Outcome'] = df['Booking Status'].map({
    'Success': 'Completed',
    'Cancelled By Driver': 'Cancelled',
    'Cancelled By Customer': 'Cancelled',
    'Incomplete': 'Incomplete'
})
```

```python
df[['Booking Value','Ride Distance','Driver Ratings','Customer Rating']] = (
    df[['Booking Value','Ride Distance','Driver Ratings','Customer Rating']]
)
```

```python
df['Month'] = df['Date'].dt.to_period('M').astype(str)
```

```python
df['Customer_Cancel'] = df['Cancelled  by Customer'].apply(lambda x: 'Yes' if x==1 else 'No')
df['Driver_Cancel'] = df['Cancelled Rides by Driver'].apply(lambda x: 'Yes' if x==1 else 'No')
```

```python
total_revenue = df['Booking Value'].sum()
```

```python
cancelled_rides = df[df['Ride Outcome']=='Cancelled'].shape[0]
total_rides = df.shape[0]

cancellation_rate = (cancelled_rides / total_rides) * 100
```

```python
vehicle_perf = df.groupby('Vehicle Type')['Booking Value'].sum().sort_values(ascending=False)
```
